```python
def emission_color(total_emission):
    if total_emission < 1000000:
        return "green"
    elif 1000000 <= total_emission < 2000000:
        return "orange"
    else:
        return "red"
```

```python
lat = list( df_new['latitude'])
lon = list(df_new['longitude'])
pop = list(df_new['population'])
city = list(df_new["city"])
```

```python
te_co2 = list(df_new['total_emissions_(metric_tonnes_co2e)'])
```

```python
df_new
```

```python
map2 = folium.Map(location=[54.891456,-10.404684], zoom_start=6, tiles = 'Mapbox Bright')
```

```python
fg = folium.FeatureGroup(name="CO2 Map")
```

```python
for lt, ln,co,ci in zip(lat,lon,te_co2,city):
    fg.add_child(folium.CircleMarker(location=[lt,ln], radius = 6,
                                     popup = "CO2 level of " +ci +" is :" +str(co) + "metric tone" ,
                                     fill_color=emission_color(co),color= 'grey', fill_opacity =0.7))
```

```python
map2.add_child(fg)
```

```python
map2.save("CO2.html")
```
